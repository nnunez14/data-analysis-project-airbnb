```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb.groupby(['neighbourhood_group']).count()
# There are 5 neighborhood groups with Manhattan appearing the most umber of times
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(['neighbourhood_group','room_type']).count()
# No, the most popular are Entire home/apt
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby('host_name').sum().sort_values('reviews_per_month')
# Michael is the busiest with 11,081 reviews. Next closest is David with 8,103.
```

```python
#Which neighorhood group has the highest average price?

air_bnb.groupby('neighbourhood_group').mean()['price'].sort_values(ascending = False)
# Manhattan has the highest average price
```

```python
# Which neighborhood group has the highest total price?

air_bnb.groupby('neighbourhood_group').sum()['price'].sort_values(ascending = False)
# Manhattan has the highest total price
```

```python
#Which top 5 hosts have the highest total price?

air_bnb.groupby('host_name').sum().sort_values('price',ascending = False)['price']
# from highest to lowest:
# 1. Sonder (NYC)         82795
# 2. Blueground           70331
# 3. Michael              66895
# 4. David                65844
# 5. Alex                 52563
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

popular = air_bnb[air_bnb['availability_365'] == 0][air_bnb['number_of_reviews'] >= 100]
popular = popular.sort_values('number_of_reviews',ascending = False)
popular
# popular.count() # To get the total number of unique hosts that fall under this criteria

# Wanda, Gurpreet Singh, Deloris, Veronica have the highest number of reviews
# There are 161 unique hosts that fall under this criteria, 162 total hosts with this criteria
```

```python
# What host has the highest total of prices and where are they located?

bougie = air_bnb.groupby('host_name').sum().sort_values('price', ascending = False).head()
print(bougie)

# Sonders at latitude 13316.25823, longitude -24198.18856
```

```python
# When did Danielle from Queens last receive a review?

danielle = air_bnb[air_bnb['host_name'] == 'Danielle'][air_bnb['neighbourhood_group'] == 'Queens'].sort_values('last_review', ascending = False)
danielle['last_review'].iloc[0]

# both line 3 and 4 work. 
# line 3 will display the table if i call the variable 'danielle'.

# line 4 will just print the most recent date

# answer = 2019-07-08
```
